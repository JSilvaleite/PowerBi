# Selecionando dados apenas de São Paulo

```python
base_covid_sp = base_covid[base_covid['UF'] == 'SP']
```

## Selecionando colunas

```python
base_covid_sp_selecionado = base_covid_sp[['regiao', 'estado', 'municipio', 'coduf', 'codmun', 'codRegiaoSaude',
       'nomeRegiaoSaude', 'data', 'semanaEpi', 'populacaoTCU2019',
       'casosAcumulado', 'casosNovos', 'obitosAcumulado', 'obitosNovos',
       'Recuperadosnovos', 'emAcompanhamentoNovos', 'interior/metropolitana', 'UF']]
```

```python
base_covid_sp_selecionado.head(5)
```

# Transformando dados

```python
base_covid_sp_selecionado['municipio'] = base_covid_sp_selecionado.municipio.fillna('0')
```

```python
base_covid_sp_selecionado['municipio'] = base_covid_sp_selecionado.municipio.fillna('0')
```

```python
base_covid_sp_selecionado['codmun'] = base_covid_sp_selecionado.codmun.fillna('0')
```

```python
base_covid_sp_selecionado['codRegiaoSaude'] = base_covid_sp_selecionado.codRegiaoSaude.fillna('0')

```

```python
base_covid_sp_selecionado['nomeRegiaoSaude'] = base_covid_sp_selecionado.nomeRegiaoSaude.fillna('0')

```

```python
base_covid_sp_selecionado = base_covid_sp_selecionado.fillna(0)
```

```python
base_covid_sp_selecionado.head()
```

# Alterando tipo de dados

```python
base_covid_sp_selecionado.dtypes
```

```python
colns = base_covid_sp_selecionado.select_dtypes(include=np.float64)
```

```python
colns.columns
```

```python
base_covid_sp_selecionado[['populacaoTCU2019', 'casosAcumulado', 'Recuperadosnovos',
       'emAcompanhamentoNovos', 'interior/metropolitana']] = base_covid_sp_selecionado[['populacaoTCU2019', 'casosAcumulado', 'Recuperadosnovos',
       'emAcompanhamentoNovos', 'interior/metropolitana']].astype(int)
```

```python
base_covid_sp_selecionado = base_covid_sp_selecionado[base_covid_sp_selecionado['municipio']!='0']
```
